# **Imports, Loads do Raw Dataset e Previsualização dos dados para entende-lo**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/DNC - Cursos/4 Curso Ciencias de Dados/1 CRISP - DM/CRISP - DM - Data Preparation na Pratica/DataPrepFinal.csv")


df.head()
```

# **Select Data - Selecionando os dados que iremos trabalhar**

```python
# Unnamed - não faz sentido
# name - não iremos trabalhar com NLP(Processamento de linguagem natural)
# category - a main category parece ser mais relevante
# goal - tem tipos de moedas diferentes, diferente do usd_goal_real que esta tudo em dolar
# pledged - iremos remover pelo mesmo motivo do goal
# usd pledged - esta com dados errados e faltantes
df.drop(columns = ["Unnamed: 0", "name", "category", "goal", "pledged", "usd pledged"], inplace = True)


df.head()
```

# **Clean Data - Vamos fazer uma limpeza nos dados**

```python
# Removendo 'USD ' da coluna 'usd_goal_real' - apply: percorre cada linha do dado
df["usd_goal_real"] = df["usd_goal_real"].apply(lambda x: x.strip('USD '))


df.head()
```

```python
# Usando o Regex + replace para remover da hora na coluna 'launched'
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex = True)


df.head()
```

```python
df.dtypes
```

```python
# Mudando para os tipos de dados corretos
df["usd_goal_real"] = df["usd_goal_real"].astype('int64')
df["deadline"] = pd.to_datetime(df["deadline"], format = '%d/%m/%y')
df["launched"] = pd.to_datetime(df["launched"], format = '%d/%m/%y')


df.dtypes
```

# **Construct Data - Criando uma nova coluna**

```python
# Criando uma nova coluna - baseando-se nas duas coluna "launched" - "deadline"
# dt: calsse de datas, days: metodo para obter apenas os dias
df["time_range"] = (abs(df["launched"] - df["deadline"])).dt.days


df.head()
```

```python
df.dtypes
```

## Fazendo o 3 import para realizar o merge

```python
df_invested = pd.read_csv("/content/drive/MyDrive/DNC - Cursos/4 Curso Ciencias de Dados/1 CRISP - DM/CRISP - DM - Data Preparation na Pratica/invested.csv")


df_invested.head()
```

```python
# PROJETO --> ID: 106144, tem mais de um registro
# Indicando que recebeu mais de um investimento de pessoas diferentes
# e para os demais é a mesma coisa

# Para fazer o merge dessa tabela com a primeira ira gerar problema pois o ID
# não é unico, no caso gera inconsistencia nos dados

# Então iremos utilizar a tecnica de ONEHOTCODE
df_invested.sort_values(by = ['ID'])
```

## ONEHOTCODE - Tranformar as variaveis categoricas em numericas, 1 = True 0 = False

```python
# Fizemos ONEHOTCODE para a coluna 'backedLocation'
# Porem a coluna 'age' ainda causara inconsistencia nos dados
# mas com ela iremos usar a tecnica de groupby + agg

# AGE:agregar usando o metodo de mean/media
# Os demais: agregar usando o metodo soma
df_invested = pd.get_dummies(df_invested, columns = ['backedLocation'], dtype = int)


df_invested.sort_values(by = ['ID'])
```

```python
# AGE:agregar usando o metodo de mean/media
# Os demais: agregar usando o metodo sum/soma
df_invested = df_invested.groupby(by = ['ID']).agg({'age': 'mean',
                                                    'backedLocation_BR': 'sum',
                                                    'backedLocation_GBK': 'sum',
                                                    'backedLocation_US': 'sum'})


df_invested.sort_values(by = ['ID'])
```

```python
df_invested.age = df_invested.age.astype('int64')


df_invested.head()
```

## Depois de usar técnicas como OneHotCode na 3 tabela, vamos fazer o merge com a 1 tabela

```python
df = df.merge(df_invested, how = 'right', on = ['ID'])
```

## Raw Dataset agora foi transformado para Dataset
